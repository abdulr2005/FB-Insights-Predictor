```python
import pandas as pd

df = pd.read_csv("Live.csv")

df
```

```python
df =df.drop(columns= ["status_id","Column1","Column2","Column3","Column4","status_published"])

df.head(5)
```

```python
df.info()
```

```python
status_type = {
    "video":0 ,
    "photo":1
}

df['status_type'] = df['status_type'].map(status_type)
df.head(5)
```

```python
df['status_type'].isnull()
```

```python
df['status_type'] = df['status_type'].fillna(df['status_type'].mode()[0])
df.info()
```

```python
df['status_type'].isnull().sum()
```

```python
df['status_type'] = df['status_type'].astype(int)
df.info()
```

clustring sec

```python
# scaler 

from sklearn.preprocessing import StandardScaler

X = df

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
#PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

```

```python
#find the best (n_clusters)

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for i in range (1,11) :
    kmeans = KMeans(n_clusters= i , init='k-means++',random_state=42)
    kmeans.fit(X_pca)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3,random_state=42)
clusters = kmeans.fit_predict(X_pca)
df['Cluster'] = clusters
```

```python
cluster_analysis = df.groupby('Cluster').mean()
print(cluster_analysis)
```

```python
df.to_csv("LIVE_after_clean_and_clus.csv")
```
